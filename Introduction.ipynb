```python
from pathlib import Path
import pandas as pd
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
Path.cwd()
```

```python
rain_data = pd.read_excel("rainfall.xlsx", sheet_name="Sheet1")  # Import Excel file
```

```python
print(rain_data.head()) # view data
```

```python
cols = rain_data[["Years", "Jan"]]
```

```python
print(cols.head())
```

```python
output_folder = "output"
os.makedirs(output_folder, exist_ok=True)  # Create an output folder if it doesn’t exist
rain_data.plot(x="Years", y="Jan", kind="line", marker="o", figsize=(8,5))
plt.xlabel("Years")
plt.ylabel("Rainfall (mm)")
plt.title("January Rainfall - Kajiado")
plt.grid()
image_path = os.path.join(output_folder, "January_rainfall.png")
plt.savefig(image_path)

print(f"Plot saved at: {image_path}")
```
